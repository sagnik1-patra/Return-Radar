```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

OUT_DIR = Path(r"C:\Users\NXTWAVE\Downloads\Return Radar")
CM_CSV  = OUT_DIR / "confusion_matrix.csv"
H5_PATH = OUT_DIR / "processed_returns.h5"   # fallback (key='holdout')

OUT_DIR.mkdir(parents=True, exist_ok=True)

def load_confusion_matrix():
    """Load cm + labels either from CSV (preferred) or rebuild from HDF5 holdout."""
    if CM_CSV.exists():
        cm_df = pd.read_csv(CM_CSV, index_col=0)
        # labels like ['0','1'] derived from columns 'pred_0','pred_1'
        labels = [c.replace("pred_", "") for c in cm_df.columns]
        ytick = [i.replace("true_", "") for i in cm_df.index]
        if labels != ytick:
            # keep them consistent if format differs
            ytick = labels
        cm = cm_df.values
        return cm, labels
    # Fallback: rebuild cm from HDF5 holdout predictions
    if H5_PATH.exists():
        try:
            holdout = pd.read_hdf(H5_PATH, key="holdout")
            y_true = holdout["returned_true"].astype(int)
            y_pred = holdout["returned_pred"].astype(int)
            # Build a 2x2 cm with labels [0,1] even if one class absent
            cm = np.zeros((2,2), dtype=int)
            for t, p in zip(y_true, y_pred):
                if t in (0,1) and p in (0,1):
                    cm[t, p] += 1
            labels = ["0","1"]
            return cm, labels
        except Exception as e:
            raise SystemExit(f"Could not rebuild confusion matrix from HDF5: {e}")
    raise SystemExit(
        "No confusion matrix found. Run training first to create "
        "'confusion_matrix.csv' or 'processed_returns.h5'."
    )

# ---- Load CM & compute accuracy
cm, labels = load_confusion_matrix()
total = cm.sum()
acc = (np.trace(cm) / total) if total else 0.0
print("Accuracy:", round(acc, 4))
print("Confusion matrix:\n", cm)

# ---- Plot & save: Accuracy bar
plt.figure()
ax = sns.barplot(x=["accuracy"], y=[acc])
ax.set_ylim(0, 1)
ax.set_title("Model Accuracy")
for c in ax.containers:
    ax.bar_label(c, fmt="%.3f")
plt.tight_layout()
acc_path = OUT_DIR / "viz_accuracy.png"
plt.savefig(acc_path, dpi=160)
plt.show()

# ---- Plot & save: Confusion Matrix (counts)
plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted"); plt.ylabel("True")
plt.title("Confusion Matrix (counts)")
plt.tight_layout()
cm_counts_path = OUT_DIR / "viz_confusion_matrix.png"
plt.savefig(cm_counts_path, dpi=160)
plt.show()

# ---- Plot & save: Confusion Matrix (row-normalized)
row_sums = cm.sum(axis=1, keepdims=True)
norm = cm / np.where(row_sums == 0, 1, row_sums)
plt.figure()
sns.heatmap(norm, annot=True, fmt=".2f", cmap="Greens",
            xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted"); plt.ylabel("True")
plt.title("Confusion Matrix (row-normalized)")
plt.tight_layout()
cm_norm_path = OUT_DIR / "viz_confusion_matrix_normalized.png"
plt.savefig(cm_norm_path, dpi=160)
plt.show()

print("Saved:")
print(" ", acc_path)
print(" ", cm_counts_path)
print(" ", cm_norm_path)
```
